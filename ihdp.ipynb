```python
from sklearn.datasets import make_regression
from sklearn.tree import DecisionTreeRegressor
from matplotlib import pyplot
X, y = ihdp[ihdp.columns[:25]],ihdp['yf']
model = DecisionTreeRegressor()
model.fit(X, y)
importance = model.feature_importances_
for i,v in enumerate(importance):
	print('x%0d, Score: %.5f' % (i+1,v))
pyplot.bar([x for x in range(len(importance))], importance)
pyplot.show()
```

```python
pyplot.bar([x for x in range(len(importance))], importance)
pyplot.title("Feature Importance for IHDP dataset using Decision Tree Regressor")
pyplot.ylabel("Feature Score")
pyplot.xlabel("Feature Number")
pyplot.show()
```

```python
Features_Importance=pd.DataFrame([ihdp.columns[:25],importance]).T
```

```python
Features_Importance.columns=['Feature Name','Score']
```
